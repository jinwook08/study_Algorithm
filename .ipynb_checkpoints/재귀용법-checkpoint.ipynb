```python
# 팩토리얼 구하기
# 입력: n
# 출력: 1부터 n까지 연속한 숫자를 곱한 값.
def factorial(n):
    fac = 1                  # 곱을 계산할 변수, 초깃값은 1
    for i in range(1, n + 1):     # 1 부터 n까지 반복( n + 1은 제외 )
        fac = fac * i                   # 곱셈 연산으로 수정
    return fac


print(factorial(1))
```
